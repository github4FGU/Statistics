```python
import matplotlib
import random
import numpy as np
import matplotlib.pyplot as plt
```

# Gauß'sche Normalverteilung #

Eine Gaußsche Glockenkurve nur mit Zufallszahlen generiert:

```python
rand = []
for j in range(100_000):
    sum = 0.0
    for i in range(200):
        x = random.random()
        sum += x
    rand.append(sum)

# print(rand)
```

```python
# example data
mu = 100.0  # mean of distribution
sigma = 4.0  # standard deviation of distribution
H = 37_000.0  # Maximale Höhe
num_bins = 100

fig, ax = plt.subplots()
# the histogram of the data
n, bins, patches = ax.hist(rand, num_bins)

# add a 'best fit' line
y = ((H / (np.sqrt(2.0 * np.pi) * sigma)) *
     np.exp(-0.5 * (1.0 / sigma * (bins - mu))**2))
ax.plot(bins, y, '--')
ax.set_xlabel('Summen')
ax.set_ylabel('Anzahl')
ax.set_title(r'Histogram of IQ: $\mu=100$, $\sigma=5$')

# Tweak spacing to prevent clipping of ylabel
fig.tight_layout()
plt.show()
```
